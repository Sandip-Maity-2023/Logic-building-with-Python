```python
import numpy as p
import matplotlib.pyplot as pl
from sklearn.linear_model import LinearRegression
```

```python
p.random.seed(0) #p.random refers to the module inside Numpy & .seed(0) function to set the seed for random number generation to 0, without seed the random numbers will be different each time the code is run.
#predictable and reproducible results

x=p.random.rand(50,1)*100 #50 random values between 0 and 10 in a 50x2 array

y=3.5*x+p.random.randn(50,1)*20 #means y is generated using a linear equation with some added noise

model=LinearRegression() #Creating an instance of the LinearRegression class from scikit-learn
model.fit(x,y) #Fitting the linear regression model to the data
y_prediction=model.predict(x) #Making predictions using the fitted model

print(y_prediction) #Printing the predicted values

#visualization
pl.figure(figsize=(10,5)) #Creating a new figure with a specified size of 10 inches by 5 inches
pl.scatter(x,y,color='blue',label='Actual Data') #Creating a scatter plot of the original data points in blue
pl.plot(x,y_prediction,color='red',linewidth=2,label='Regression Line') #Creating a scatter plot of the predicted data points in red

pl.title('Linear Regression on random dataset') #Setting the title of the plot
pl.xlabel('X-axis') #Labeling the x-axis
pl.ylabel('Y-axis') #Labeling the y-axis
pl.legend() #Adding a legend to the plot to differentiate between actual data and regression line

pl.grid(True) #Adding a grid to the plot for better readability
pl.show() #Displaying the plot
```

```python
print("Slope(Coefficient):",model.coef_[0][0]) #Printing the slope (coefficient) of the regression line & [0][0] to access the first element of the 2D array
print("Intercept:",model.intercept_[0]) #Printing the y-intercept of the regression line
```
