Implementation of gradient descent in Linear regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

x,y=make_regression(n_samples=100,n_features=1,noise=10,random_state=42) 

#1st 5 rows of feature matrix x and target vector y
print("x[:5]",x[:5])
print("y[:5]",y[:5]) 

y=y.reshape(-1,1) #mean -1 indicates that the number of rows is inferred from the length of the array and 1 indicates that there is one column
print("reshaped y[:5]",y[:5])

m=x.shape[0]  #mean shape[0] gives the number of rows in x which is equal to the number of samples

x_b=np.c_[np.ones((m,1)),x] #adding a bias term (intercept) to the feature matrix x by concatenating a column of ones to x
print("x_b[:5]",x_b[:5])

theta=np.array([[2.0],[3.0]]) #initializing the parameters (weights) theta with some arbitrary values

plt.figure(figsize=(10,6))
plt.scatter(x,y,color='blue',label='Data Points')
plt.plot(x,x_b.dot(theta),color='green',label='Initial Regression Line') #x_b.dot(theta) computes the predicted values using the initial parameters theta

plt.xlabel('Feature X')
plt.ylabel('Target Y')
plt.title('Initial Regression Line with Data Points without Gradient Descent')
plt.legend() 
plt.show()
```

```python
learning_rate = 0.1  #step size for each iteration
n_iterations = 100

for _ in range(n_iterations):

    y_predicted = x_b.dot(theta)
    gradients = (2 / m) * x_b.T.dot(y_predicted - y) #.T is used to transpose the matrix x_b,.dot() is used to perform matrix multiplication
    theta -= learning_rate * gradients

plt.figure(figsize=(10, 5))

plt.scatter(x, y, color="blue", label="Actual Data")
plt.plot(x, x_b.dot(theta), color="red", label="Optimized Line (With GD)")

plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Linear Regression With Gradient Descent")
plt.legend()
plt.show()
```
