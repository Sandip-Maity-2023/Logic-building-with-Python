```python
import matplotlib.pyplot as plt #used for plotting and visualization
import pandas as pd #used for data manipulation and analysis
import numpy as np #used for numerical operations
```

```python
df = pd.read_csv('placement[1].csv')
```

```python
df.head() # Display the first few rows of the DataFrame
```

```python
plt.scatter(df['cgpa'],df['package']) # Scatter plot of CGPA vs Package
plt.xlabel('CGPA') # Label for x-axis
plt.ylabel('Package(in lpa)') # Label for y-axis
plt.title('CGPA vs Package') # Title of the plot
```

```python
X = df.iloc[:,0:1] # Selects all rows and the first column of the DataFrame.
#X = df.iloc[:,:-1] # Selects all rows and all columns except the last one.
y = df.iloc[:,-1] #: (before the comma): Selects all rows in the DataFrame.
#-1 (after the comma): Refers to the last column of the DataFrame. Negative indexing (-1) is a common feature in Python to access elements from the end.
#x = df['cgpa']
#y = df['package']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
# Splitting the dataset into training and testing sets
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
#test_size=0.2 means 20% of the data will be used for testing and 80% for training.
#random_state=2 ensures that the split is reproducible, meaning you will get the same split every time you run the code.
```

```python
from sklearn.linear_model import LinearRegression # Importing the LinearRegression class from sklearn.linear_model
```

```python
lr = LinearRegression() # Creating an instance of the LinearRegression class
#fit() method is used to train the model on the training data.
```

```python
lr.fit(X_train,y_train) # The fit method takes the training data (X_train and y_train) as input and learns the relationship between the features (X) and the target variable (y).
# The model learns the coefficients (weights) that best fit the training data.
```

```python
X_test # Display the test features (X_test) to see the data that will be used for predictions
y_test # Display the test target variable (y_test) to see the actual values for comparison
```

```python
y_test
```

```python
lr.predict(X_test.iloc[0].values.reshape(1,1))
 #Refers to the first row (since indexing starts at 0)
 # Predicting the package for the first test sample
# The reshape(1,1) method is used to convert the input into a 2D array with one row and one column, as required by the predict method.
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(X_train,lr.predict(X_train),color='red') # Plotting the regression line
plt.scatter(X_test,y_test,color='green') # Scatter plot of test data points
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

```python
coefficients = lr.coef_ # Coefficients of the linear regression model that represent the slope of the line.
```

```python
b = lr.intercept_ # Intercept of the linear regression model that represents the y-intercept of the line.
# The intercept is the value of the target variable (y) when all features (X) are equal to zero.
```
