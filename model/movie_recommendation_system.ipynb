Movie Recommendation System

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

path='movie.tsv'

columns=['user_id','item_id','rating','timestamp']
df=pd.read_csv(path,sep='\t',names=columns)         #names are predefined

df.head()
```

```python
titles=pd.read_csv('movie.csv')
titles.head(4)
```

```python
titles.tail(4)
```

```python
data=pd.merge(df,titles,on='item_id')  #merge two dataframes on item_id
data.head()
```

```python
data.tail()
```

```python
data.groupby('titles')['rating'].mean().sort_values(ascending=False).head()
```
