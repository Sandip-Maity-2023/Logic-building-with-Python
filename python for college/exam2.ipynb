```python
matrix_4x4 = np.random.randint(1, 21, (4, 4))

print("Matrix:\n", matrix_4x4)
print("Sum:", matrix_4x4.sum())
print("Mean:", matrix_4x4.mean())
print("Max:", matrix_4x4.max())
print("Min:", matrix_4x4.min())

print("Row-wise sum:", matrix_4x4.sum(axis=1))
print("Column-wise sum:", matrix_4x4.sum(axis=0))
print(matrix_4x4.sum(axis=1))
```

```python
A = np.random.randint(1, 5, (3, 3))
B = np.random.randint(1, 5, (3, 3))

dot_product = np.dot(A, B)

matrix_2x3 = np.array([[1, 2, 3], [4, 5, 6]])
transpose_matrix = matrix_2x3.T

print(dot_product)
print(transpose_matrix)
```

```python
rand_float = np.random.rand(3, 3)
print(rand_float)

large_array = np.random.uniform(0, 100, 1000)

print("Standard Deviation:", np.std(large_array))
print("Variance:", np.var(large_array))
```

```python
M1 = np.array([[1, 2], [3, 4]])
M2 = np.array([[5, 6], [7, 8]])

print(M1 + M2)
print(M1 - M2)
print(M1 * M2)

matrix = np.array([[1, 2, 3], [4, 5, 6]])
print(matrix.T)

square_matrix = np.array([[2, 3], [1, 4]])
determinant = np.linalg.det(square_matrix)

print("Determinant:", determinant)



invertible_matrix = np.array([[4, 7], [2, 6]])
inverse_matrix = np.linalg.inv(invertible_matrix)

print("Inverse:\n", inverse_matrix)



eig_matrix = np.array([[2, 0], [0, 3]])
eigenvalues, eigenvectors = np.linalg.eig(eig_matrix)

print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)

```

```python
arr = np.array([5, 2, 9, 1, 7])
sorted_arr = np.sort(arr)

print(sorted_arr)

descending_arr = np.sort(arr)[::-1]
print(descending_arr)

indices = np.where(arr > 5)
print(indices)

matrix = np.array([[3, 1, 2],
                   [9, 7, 8],
                   [6, 5, 4]])

# Sort each row
row_sorted = np.sort(matrix, axis=1)

# Sort each column
column_sorted = np.sort(matrix, axis=0)

print("Row-wise sorted:\n", row_sorted)
print("Column-wise sorted:\n", column_sorted)

```
