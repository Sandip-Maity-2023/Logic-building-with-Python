```python
#bucket Sort

def bucket_sort(arr):
    n=len(arr)                      #no of buckets
    buckets=[[] for _ in range(n)]  #create n empty buckets

    for val in arr:                 #distribute input array values into buckets
        index=int(n*val)
        buckets[index].append(val)

    for b in buckets:        #sort individual buckets   
        b.sort()

    result=[]
    for b in buckets:   #concatenate sorted buckets 
        result.extend(b)

    return result

#arr=[0.43,0.00,4.5,0.433,0.01,0.51]
#- Youre assuming all values are fractions between 0 and 1.
arr=[0.43,0.00,0.5,0.433,0.01,0.51]
print("Sorted array:",bucket_sort(arr))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def f(x, y):
     return x**2 + y**2 + 3*x + 4*y + 5

x = np.linspace(-3, 1, 400)
y = np.linspace(-3, 1, 400)
x, y = np.meshgrid(x, y)
z = f(x, y)

min_x = -1.5
min_y = -2.0
min_z = f(min_x, min_y)

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d') #create 3d plot and 111 used to specify single plot(1 row, 1 column, 1st plot)

ax.plot_surface(x, y, z, cmap='viridis', alpha=0.8)  #cmap=viridis used for color scheme,alpha for transparency
ax.scatter(min_x, min_y, min_z, color='red', s=100,
               label="Min at (min_x, min_y, min_z)")

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('f(x, y)')

ax.legend()
plt.show()
```

```python
# Generate a signal composed of two sinusoidal components (5 Hz and 10 Hz) and plot both the time-domain signal and its frequency-domain (Fourier Transform) representation.

import numpy as np
import matplotlib.pyplot as plt

#SIGNAL GENERATION
fs=1000 #sampling frequency 1000 samples per second
t=np.linspace(0,1,fs,endpoint=False) #time vector t from 0 to 1 second with fs equally spaced points(so 1000 samples)
#endpoint=False ensures the last value is just before 1,not exactly 1

signal1=np.sin(2*np.pi*5*t) #5 Hz sine wave
signal2=np.sin(2*np.pi*10*t) #10 Hz sine wave
signal=signal1+signal2 #composite signal

#FFT COMPUTATION
N=len(signal)
vals=np.fft.fft(signal)/N #compute Fourier Transform
freq=np.fft.fftfreq(N,1/fs) #generate the corresponding bins for the First Fourier Transform values
# len(vals)=number of samples in the signal
# 1/fs=sampling interval(0.001 seconds)

#PLOTTING
plt.figure(figsize=(12,6))
plt.subplot(2,1,1) #Divides the figure into 2 rows, 1 column, first plot
plt.plot(t,signal) #plots the signal amplitude vs time
plt.title("Time-Domain Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")


plt.subplot(2,1,2) #second plot
plt.stem(freq[:N//2], 2*np.abs(vals[:N//2]))
#creates a stem plot(vertical lines with markers) for frequency vs magnitude of Fourier Transform
# freq[:len(freq)//2] only plots the positive half of frequencies(FFT is symmetric around 0)
# np.abs(vals[:len(freq)//2]) gets the magnitude of the FFT values for positive frequencies

plt.title("Frequency-Domain Representation (Fourier Transform)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.xlim(0,50) #limit x-axis to 0-50 Hz for better visibility
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt



# Sampling parameters
fs = 1000                 # Sampling frequency (Hz)
t = np.linspace(0, 1, fs, endpoint=False)  # Time vector (1 second)

# Signal generation
signal1 = np.sin(2 * np.pi * 5 * t)   # 5 Hz sine wave
signal2 = np.sin(2 * np.pi * 10 * t)  # 10 Hz sine wave
signal = signal1 + signal2            # Composite signal

# Fourier Transform
N = len(signal)
fft_vals = np.fft.fft(signal) / N     # Normalized FFT
freq = np.fft.fftfreq(N, 1 / fs)      # Frequency bins

# Plotting
plt.figure(figsize=(12, 6))

# Time-domain plot
plt.subplot(2, 1, 1)
plt.plot(t, signal)
plt.title("Time-Domain Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")

# Frequency-domain plot (single-sided spectrum)
plt.subplot(2, 1, 2)
plt.stem(freq[:N // 2], 2 * np.abs(fft_vals[:N // 2]))
plt.title("Frequency-Domain Representation (Fourier Transform)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.xlim(0, 50)

plt.tight_layout()
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

# Sampling parameters
fs = 1000          # Sampling frequency
t = np.linspace(0, 1, fs, endpoint=False)

# Signal components
signal1 = np.sin(2 * np.pi * 5 * t)     # 5 Hz signal
signal2 = np.sin(2 * np.pi * 10 * t)    # 10 Hz signal

# Combined signal
signal = signal1 + signal2

# Fourier Transform
fft_vals = np.fft.fft(signal)
freq = np.fft.fftfreq(len(fft_vals), 1/fs)

# Plot time-domain signal
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(t, signal)
plt.title("Time Domain Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")

# Plot frequency-domain signal
plt.subplot(2, 1, 2)
plt.stem(freq[:len(freq)//2], np.abs(fft_vals[:len(freq)//2]))
plt.title("Frequency Domain Signal")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")

plt.tight_layout()
plt.show()
```

```python
class Animal:
    # attribute and method of the parent class
    name = ""

    def eat(self):
        print("I can eat")

# inherit from Animal
class Dog(Animal):
    # new method in subclass
    def display(self):
        # access name attribute of superclass using self
        print("My name is", self.name)


# create an object of the subclass
labrador = Dog()
# access superclass attribute and method
labrador.name = "Rohu"
labrador.eat()
# call subclass method
labrador.display()
```

```python
class Calculator:

    def add(self,a,b):
        return a+b
    def multi(self,a,b):
        return a*b
    def sub(self,a,b):
        return a-b
    def div(self,a,b):
        if b==0:
            return "Division by zero not allowed"
        return a/b
    
calc=Calculator()
print(calc.add(5,3))
print(calc.multi(4,7))
print(calc.div(10,2))
print(calc.div(10,0))
#- n Python, instance methods must include self as the first parameter

```

```python
class Employee:
    def __init__(self, name, basic):
        self.name = name
        self.basic = basic

    def gross_salary(self):
        hra = 0.20 * self.basic
        da = 0.10 * self.basic
        return self.basic + hra + da

emp = Employee("Amit", 30000)
print("Gross Salary:", emp.gross_salary())
```

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Amit", "Sara", "Amit", "John", None, "Meera"],
    "Age": [21, 22, 21, None, 23, 22],
    "City": ["Delhi", "Mumbai", "Delhi", "Kolkata", "Chennai", None]
}
df = pd.DataFrame(data)
print(df)

print(df.isnull().sum())

print(df.drop_duplicates())

df["Name"].fillna("Unknown", inplace=True)
print(df)

# Ensure 'Age' is numeric before computing mean, then fill missing values with the mean
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
df["Age"].fillna(df["Age"].mean(), inplace=True)
print(df)

print(df.dropna(subset=["City"]))



```
