```python
import numpy as np

A=np.array([[7,2,1],
            [4,5,2],
            [5,2,-1]])
B=np.array([8,10,32])
roots = np.linalg.solve(A, B)
print("Roots (x,y,z):", roots)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,10,100)
y=np.sin(x)+np.cos(x)

plt.plot(x,y,label='y=sin(x)+cos(x)')
plt.title('Signal plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.grid(True)
plt.show()
```

```python
import numpy as np

A=np.array([[2,3,1],
           [4,5,6],
           [7,8,9]])

B=np.array([[1,0,2],
            [3,4,5],
            [6,7,8]])
dot_pro=np.dot(A,B)
print("Dot Product:\n",dot_pro)

transpose=dot_pro.T
print("Transpose:\n",transpose)

inverse=np.linalg.inv(A)
print("\nInverse of A:\n",inverse)

det=np.linalg.det(A)
print("\nDeterminant of A:",det)

eigen_values,eigen_vectors=np.linalg.eig(A)
print("\nEigen Values:\n",eigen_values)
print("\nEigen Vectors:\n",eigen_vectors)
```

```python
import numpy as np
from scipy import stats

data=np.random.randint(0,100,1000)
mean=np.mean(data)
median=np.median(data)
mode=stats.mode(data,keepdims=False).mode
std=np.std(data)
variance=np.var(data)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Standard Deviation: {std}")
print(f"variance: {variance}")
```
