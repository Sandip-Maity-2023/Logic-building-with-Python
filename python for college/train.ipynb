```python
import pandas as pd
from sklearn.model_selection import train_test_split

data=pd.read_csv('10heart.csv')
print(data.head())

x=data.drop('AHD',axis=1)
y=data['AHD']  

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
numerical_cols=["Age","RestBP","Chol","MaxHR"]

mean_values = x_train[numerical_cols].mean()
median_values = x_train[numerical_cols].median()
mode_values = x_train[numerical_cols].mode().iloc[0]

print("Mean values:\n", mean_values)
print("Median values:\n", median_values)
print("Mode values:\n", mode_values)
```
