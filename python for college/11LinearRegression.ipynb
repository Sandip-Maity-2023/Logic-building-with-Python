```python
import numpy as np;
import matplotlib.pyplot as plt

#x=np.array([1,2,3,4,5,6]) #predictor x and target y
#y=np.array([2,4,5,4,5,7])

#np.random.seed(0) #means in simple words that every time you run the code, i will get the same random numbers
x=np.random.rand(50) * 20 #random values between 0 and 20
y=3*x + 5+ np.random.randn(50) #linear relation with some noise
#y=mx+c
x_mean=np.mean(x)
y_mean=np.mean(y)

#slope(m)

numerator=np.sum((x-x_mean)*(y-y_mean))
denominator=np.sum((x-x_mean)**2)
m=numerator/denominator

#intercept(c)
c=y_mean-m*x_mean

print(f"Slope: {m}, Intercept: {c}")

#predicted values 
y_pred=m*x+c      #applies to each x value 

#plotting
plt.scatter(x,y,color='blue',label='Actual Data')
plt.plot(x,y_pred,color='red',label='Regression Line')
plt.xlabel('Predictor (x)')
plt.ylabel('Target (y)')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()

MAE = np.mean(np.abs(y - y_pred))

MSE = np.mean((y - y_pred) ** 2)

RMSE = np.sqrt(MSE)

print("MAE :", MAE)
print("MSE :", MSE)
print("RMSE:", RMSE)


```
